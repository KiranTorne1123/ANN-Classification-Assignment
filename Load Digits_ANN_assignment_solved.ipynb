IMPORT NECESSARY LIBRARIES :-

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from ucimlrepo import fetch_ucirepo 
print('Importing is done')
```

DEFINE THE COLUMN NAMES BEFORE LOADING THE DATASET:-

```python
# Define the column names based on the dataset description
column_names = [f'pixel_{i}' for i in range(64)] + ['digit']
column_names
```

LOAD THE DATASETS:-

```python
# Load the training data
train_data = pd.read_csv('optdigits.tra', header=None, names=column_names)
train_data
```

```python
# Load the test data
test_data = pd.read_csv('optdigits.tes', header=None, names=column_names)
test_data
```

MERGE THE DATASET :-

```python
# Combine train and test data if needed
df = pd.concat([train_data, test_data], ignore_index=True)
```

SPLITTING THE DATASET INTO TRAINING (x) & TESTING (y) DATA :-

```python
# Split the dataset into features and labels
x = df.iloc[:, :-1].values
x
```

```python
y = df.iloc[:, -1].values
y
```

```python
x.shape, y.shape
```

FEATURE SCALING OF THE DATA :-

```python
# we import additional libraries

from sklearn.preprocessing import StandardScaler
print('Importing successfully done')
```

```python
# Normalize the feature values
sc = StandardScaler()
sc
```

```python
x = sc.fit_transform(x)
```

APPLYING TRAIN - TEST SPLIT :-

```python
# we import additional libraries

from sklearn.model_selection import train_test_split
print('Importing successfully done')
```

```python
# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=6)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

LOAD THE MODEL:-

```python
# We import necessary libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
print("Importing successfully done")
```

CREATING SEQUENTIAL MODEL (ANN)

```python
# Initialize the ANN
model = Sequential()
model
```

```python
# ASSIGNING HIDDEN LAYERS AND THE DROPOUT TO THE DATA

# Add input layer and first hidden layer
model.add(Dense(units=64,kernel_initializer = 'normal', activation='relu', input_shape=(64,)))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))

# Add second hidden layer
model.add(Dense(units=32,kernel_initializer = 'normal', activation='relu'))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))

# Add second hidden layer
model.add(Dense(units=16,kernel_initializer = 'normal', activation='relu'))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))

# Add second hidden layer
model.add(Dense(units=64,kernel_initializer = 'normal', activation='relu'))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))
```

```python
# Add output layer
model.add(Dense(units=10, activation='softmax'))  # 10 classes for the digits 0-9
```

```python
# Compile the ANN
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Model Summary
model.summary()
```

Train the ANN

```python
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size=32, epochs=300)
```

```python
# Define needed variables
tr_loss = history.history['loss']
val_loss = history.history['val_loss']
index_loss = np.argmin(val_loss)
val_lowest = val_loss[index_loss]

Epochs = [i+1 for i in range(len(tr_loss))]
loss_label = f'best epoch= {str(index_loss + 1)}'

# Plot training history
plt.figure(figsize= (20, 8))
plt.style.use('fivethirtyeight')

plt.plot(Epochs, tr_loss, 'r', label= 'Training loss')
plt.plot(Epochs, val_loss, 'g', label= 'Validation loss')
plt.scatter(index_loss + 1, val_lowest, s= 150, c= 'blue', label= loss_label)
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout
plt.show()
```

```python
# Accuracy score:

val_accuracy = np.max(history.history['val_accuracy'])
print("\n%s: %.2f%%" % ('val_accuracy', val_accuracy*100))
```

```python
# The accuracy observed is 97.33%
```
