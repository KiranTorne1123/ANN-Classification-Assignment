# Splitting the Dataset into Training (x) and Testing (y) Data

```python
x = df.drop(columns=['LUNG_CANCER']).values
x
```

```python
y = df['LUNG_CANCER'].values
y
```

```python
x.shape, y.shape
```

# Performing train test split

```python
# we import additional libraries

from sklearn.model_selection import train_test_split
print('Importing successfully done')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=6)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Feature Scaling for the data

```python
# we import additional libraries

from sklearn.preprocessing import StandardScaler
print('Importing successfully done')
```

```python
sc = StandardScaler()
sc
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Loading the model:-

```python
# We import necessary libraries

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
print('Importing successfully done')
```

```python
# Creating Sequential Model
```

```python
# Initialising the ANN

model = Sequential()
model
```

```python
# input_dim - number of columns of the dataset

# units - number of outputs to be fed to the next layer, if any

# activation - activation function which is ReLU in this case

# kernel_initializer - the way in which weights should be provided to an ANN
```

```python
# Adding the input layer and the first hidden layer
model.add(Dense(units = 32, kernel_initializer = 'normal', activation = 'relu', input_dim = 15))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))

# Adding the second hidden layer
model.add(Dense(units = 16, kernel_initializer = 'normal', activation = 'relu'))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))

# Adding the third hidden layer
model.add(Dense(units = 8, kernel_initializer = 'normal', activation = 'relu'))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))

# Adding the fourth hidden layer
model.add(Dense(units = 4, kernel_initializer = 'normal', activation = 'relu'))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))

# Adding the fourth hidden layer
model.add(Dense(units = 2, kernel_initializer = 'normal', activation = 'relu'))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))
```

```python
# Adding the output layer
model.add(Dense(units = 1, kernel_initializer = 'normal', activation = 'sigmoid'))
```

```python
#Compiling ANN

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Optimizer is chosen as adam for gradient descent.

# Binary_crossentropy is the loss function used.

# Batch size defines number of samples that going to be propagated through the network.

# An Epoch is a complete pass through all the training data.
```

```python
# Model Summary

model.summary()
```

```python
# Fitting the ANN to the Training set

history = model.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size=100, epochs=500)
```

```python
# Define needed variables
tr_loss = history.history['loss']
val_loss = history.history['val_loss']
index_loss = np.argmin(val_loss)
val_lowest = val_loss[index_loss]

Epochs = [i+1 for i in range(len(tr_loss))]
loss_label = f'best epoch= {str(index_loss + 1)}'

# Plot training history
plt.figure(figsize= (20, 8))
plt.style.use('fivethirtyeight')

plt.plot(Epochs, tr_loss, 'r', label= 'Training loss')
plt.plot(Epochs, val_loss, 'g', label= 'Validation loss')
plt.scatter(index_loss + 1, val_lowest, s= 150, c= 'blue', label= loss_label)
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout
plt.show()
```

```python
# Accuracy score:

val_accuracy = np.max(history.history['val_accuracy'])
print("\n%s: %.2f%%" % ('val_accuracy', val_accuracy*100))
```

```python
# The accuracy is about 88% using ANN.
```

```python
# Now testing new Model prediction by providing Test Data set

y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
y_pred
```
