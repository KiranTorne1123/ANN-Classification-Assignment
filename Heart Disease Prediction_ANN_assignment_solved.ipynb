```python
# 2)Bar plot of Categorical Feature Analysis: 

# Replace values in 'Heart Disease' column
df['Heart Disease'] = df['Heart Disease'].replace({0: 'NO', 1: 'YES'})

# Define categorical columns
categorical_columns = ['Sex', 'Chest pain type', 'EKG results', 'Exercise angina', 
                       'Slope of ST', 'Number of vessels fluro', 'Thallium', 'Heart Disease']

# 1. Frequency Distribution
for column in categorical_columns:
    if column != 'Heart Disease':  # Skip the target variable in frequency distribution
        print(df[column].value_counts())
        print()

# 2. Relationship with Target Variable (using chi-square test)
from scipy.stats import chi2_contingency
for column in categorical_columns[:-1]:  # Exclude 'Heart Disease'
    contingency_table = pd.crosstab(df[column], df['Heart Disease'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f'Chi-square test for {column}:')
    print(f'   Chi2 value = {chi2}')
    print(f'   p-value = {p}')
    print()

# 3. Visualization of relationships
plt.figure(figsize=(14, 10))

for i, column in enumerate(categorical_columns[:-1]):  # Exclude 'Heart Disease' from iteration
    plt.subplot(3, 3, i + 1)
    sns.countplot(x=column, hue='Heart Disease', data=df)
    plt.title(f'{column} vs Heart Disease')

plt.tight_layout()
plt.show()
```

SPLITTING THE DATASET INTO TRAINING (x) & TESTING (y) DATA :-

```python
x = df.drop(columns=['Heart Disease'],axis = 1).values
x
```

```python
y = df['Heart Disease'].values
y
```

```python
x.shape, y.shape
```

APPLYING TRAIN - TEST SPLIT :-

```python
# we import additional libraries

from sklearn.model_selection import train_test_split
print('Importing successfully done')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=6)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

FEATURE SCALING OF THE DATA :-

```python
# we import additional libraries

from sklearn.preprocessing import StandardScaler
print('Importing successfully done')
```

```python
sc = StandardScaler()
sc
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

LOADING THE MODEL :-

```python
# We import necessary libraries

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
print('Importing successfully done')
```

```python
# Creating Sequential Model (ANN)
```

```python
# Initialising the ANN

model = Sequential()
model
```

```python
# ASSIGNING HIDDEN LAYERS AND THE DROPOUT TO THE DATA

# Adding the input layer and the first hidden layer
model.add(Dense(units = 32, kernel_initializer = 'normal', activation = 'relu', input_dim = 13))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))

# Adding the second hidden layer
model.add(Dense(units = 16, kernel_initializer = 'normal', activation = 'relu'))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))

# Adding the third hidden layer
model.add(Dense(units = 8, kernel_initializer = 'normal', activation = 'relu'))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))

# Adding the fourth hidden layer
model.add(Dense(units = 4, kernel_initializer = 'normal', activation = 'relu'))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))

# Adding the fifth hidden layer
model.add(Dense(units = 2, kernel_initializer = 'normal', activation = 'relu'))
# Adding dropout to prevent overfitting
model.add(Dropout(0.5))
```

```python
# Adding the output layer

model.add(Dense(units = 1, kernel_initializer = 'normal', activation = 'sigmoid'))
```

```python
#Compiling ANN

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Model Summary

model.summary()
```

```python
# Fitting the ANN to the Training set

history = model.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size=100, epochs=300)
```

```python
# Define needed variables
tr_loss = history.history['loss']
val_loss = history.history['val_loss']
index_loss = np.argmin(val_loss)
val_lowest = val_loss[index_loss]

Epochs = [i+1 for i in range(len(tr_loss))]
loss_label = f'best epoch= {str(index_loss + 1)}'

# Plot training history
plt.figure(figsize= (20, 8))
plt.style.use('fivethirtyeight')

plt.plot(Epochs, tr_loss, 'r', label= 'Training loss')
plt.plot(Epochs, val_loss, 'g', label= 'Validation loss')
plt.scatter(index_loss + 1, val_lowest, s= 150, c= 'blue', label= loss_label)
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout
plt.show()
```

```python
# Accuracy score:

val_accuracy = np.max(history.history['val_accuracy'])
print("\n%s: %.2f%%" % ('val_accuracy', val_accuracy*100))
```

```python
# The accuracy is about 87.04% using  Sequential ANN model.
```

```python
# Now testing new Model prediction by providing Test Data set

y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
y_pred
```
